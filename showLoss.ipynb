```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
jsonFile = json.load(open('/data/gengshan/model_idtest.json', 'r'))
dictTrainLoss = jsonFile['loss_history']
dictTrainLoss = sorted(dictTrainLoss.items(), key = lambda x: int(x[0]))
dictValLoss = jsonFile['val_loss_history']
dictValLoss = sorted(dictValLoss.items(), key = lambda x: int(x[0]))
```

```python
print(dictTrainLoss)
dataTrainShow = np.zeros((2, len(dictTrainLoss)))
ix = 0
for it in dictTrainLoss:
    dataTrainShow[0, ix] = it[0]
    dataTrainShow[1, ix] = it[1]
    ix += 1
```

```python
dictValLoss = dictValLoss
print(dictValLoss)
dataValShow = np.zeros((2, len(dictValLoss)))
ix = 0
for it in dictValLoss:
    dataValShow[0, ix] = it[0]
    dataValShow[1, ix] = it[1]
    ix += 1
```

```python
valHandler = plt.plot(dataValShow[0], dataValShow[1], label= 'val')
trainHandler = plt.plot(dataTrainShow[0], dataTrainShow[1], label = 'train')
plt.ylabel('loss')
plt.xlabel('iter')
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
plt.savefig('/home/gengshan/train-val_loss.jpg', format='jpg', dpi = 1000, verbose = True)
```

```python
with open('train_log/log_8_2_16_50', 'r') as f:
    allLines = f.readlines()
    
loss = []
trainLoss = []
valLoss = []
tinyIter = []
largeIter = []

for it, line in enumerate(allLines):
    if ', loss = ' in line:
        dataPair = line.split('Iteration ')[-1].split(', loss = ')
        tinyIter.append(int(dataPair[0]))
        loss.append(float(dataPair[1])/20)
    if 'Testing net (#0)' in line:
        largeIter.append(int(line.split('Iteration ')[-1].split(', Test')[0]))
        trainLoss.append(float(allLines[it+2].split('_loss = ')[-1][:6]))
        valLoss.append(float(allLines[it+5].split('_loss = ')[-1][:7]))
assert(len(tinyIter) == len(loss))
assert(len(trainLoss) == len(valLoss) == len(largeIter))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.plot(tinyIter, loss, label = 'train')
plt.plot(largeIter, trainLoss, label = 'train')
plt.plot(largeIter, valLoss, label = 'val')
plt.xlabel('iter#')
plt.ylabel('loss')
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0.)
```
